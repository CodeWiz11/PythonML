```python
from PIL import Image
import matplotlib.pyplot as plt
image_path = 'new.png'  
image = Image.open(image_path)
image_gray = image.convert('RGB')
histogram = image_gray.histogram()
hist_r = histogram[0:256]
hist_g = histogram[256:512]
hist_b = histogram[512:768]
plt.figure(figsize=(8, 6))
plt.title('Histogram of Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.xlim(0, 256)
plt.plot(hist_r, color='red', label='Red Channel')
plt.plot(hist_g, color='green', label='Green Channel')
plt.plot(hist_b, color='blue', label='Blue Channel')
plt.legend()
plt.grid()
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = 'ferrari.JPG'
image = Image.open(image_path)

# Convert the image to RGB mode to ensure it has three channels
image_rgb = image.convert('RGB')

# Split the image into its RGB channels
r, g, b = image_rgb.split()

hist_r = r.histogram()
hist_g = g.histogram()
hist_b = b.histogram()

plt.figure(figsize=(8, 6))
plt.title('Histogram of Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.xlim(0, 256)
plt.plot(hist_r, color='red', label='Red Channel')
plt.plot(hist_g, color='green', label='Green Channel')
plt.plot(hist_b, color='blue', label='Blue Channel')
plt.legend()
plt.grid()
plt.show()
```
