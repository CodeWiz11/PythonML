# Humpty Dumpty

```python
def Humpty_Dumpty(x):
    if(x%3 == 0 and x%5==0):
        return 'Humpty_Dumpty'
    elif(x%3 == 0):
        return 'Humpty'
    elif(x%5 == 0):
        return 'Dumpty'
    else:
        return x
var = Humpty_Dumpty(2)
print(var)
```

```python
j = 0
def myFun(x):
    global j
    for i in range (0,x+1):
        if(i%3==0 or i%5==0):
            j += i
parameter = int(input('Enter parameter'))
myFun(parameter)
print('Sum is : ' + str(j))
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
read = pd.read_csv('data_line.txt',header = None)
```

```python
read
```

```python
plt.scatter(read[0],read[1])
x = np.array(read[0]).reshape((-1,1))
y = read[1]
plt.xlabel('IV')
plt.ylabel('DV')
plt.title('Linear Regression')
```

```python
from sklearn.linear_model import LinearRegression
```

```python

SLR = LinearRegression()
SLR.fit(x,y)
pred = SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,pred)
plt.plot(x,pred,color='Red',marker = '*')
print("MSE:",mse)
print("Intercept",SLR.intercept_)
print("Slope:",SLR.coef_)
```
